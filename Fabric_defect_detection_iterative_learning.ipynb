# synthetic images generation for testing

```python
import numpy as np
from PIL import Image, ImageDraw
import os

# Define the folder to save images
image_folder = "/content/drive/MyDrive/synthetic_hole_images"
os.makedirs(image_folder, exist_ok=True)

# Generate 10 synthetic images
for i in range(1, 11):
    img_size = (512, 512)  # Image size
    bg_color = (200, 200, 200)  # Gray background
    hole_color = (0, 0, 0)  # Black hole

    # Create blank image
    img = Image.new("RGB", img_size, bg_color)
    draw = ImageDraw.Draw(img)

    # Generate a random small hole
    hole_radius = np.random.randint(10, 30)  # Smaller hole size
    hole_x = np.random.randint(hole_radius, img_size[0] - hole_radius)
    hole_y = np.random.randint(hole_radius, img_size[1] - hole_radius)

    # Draw the hole (circle)
    draw.ellipse(
        (hole_x - hole_radius, hole_y - hole_radius,
         hole_x + hole_radius, hole_y + hole_radius),
        fill=hole_color
    )

    # Save the image
    img_path = os.path.join(image_folder, f"synthetic_{i}.jpg")
    img.save(img_path)

    print(f"✅ Saved: {img_path}")
```

```python
from ultralytics import YOLO
from PIL import Image, ImageDraw, ImageFont
import os
from IPython.display import display

# Load the trained YOLO model
model = YOLO("/content/drive/MyDrive/YOLO_training/exp1/weights/best.pt")

# Define the image folder
image_folder = "/content/drive/MyDrive/synthetic_hole_images"

# List of 10 test images
image_files = [f"synthetic_{i}.jpg" for i in range(1, 11)]

# Loop through each image
for image_file in image_files:
    img_path = os.path.join(image_folder, image_file)

    # Perform inference (detect objects)
    results = model(img_path)

    # Open the image
    img = Image.open(img_path)
    draw = ImageDraw.Draw(img)

    # Flag to check if a hole is detected
    hole_detected = False

    # Iterate over detected objects
    for result in results:
        for box in result.boxes:
            class_id = int(box.cls[0].item())  # Get class ID
            conf = box.conf[0].item()  # Confidence score

            # Check if the detected class is 'hole' (assuming class 0 is 'hole')
            if class_id == 0:
                hole_detected = True

                # Extract bounding box coordinates
                x1, y1, x2, y2 = map(int, box.xyxy[0])

                # Draw the bounding box
                draw.rectangle([x1, y1, x2, y2], outline="red", width=3)

                # Add label (confidence score)
                label = f"Hole: {conf:.2f}"
                draw.text((x1, y1 - 10), label, fill="red")

    # Resize for better display
    img_resized = img.resize((480, 480))  # Resize image to 480x480

    # Display the image
    display(img_resized)

    # Print result below the image
    if hole_detected:
        print("🛑 **Defect (hole) detected!**")
    else:
        print("✅ No defect detected.")

    print("-" * 50)  # Separator for clarity
```

```python
import os

# List of label directories
label_dirs = [
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/train/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/valid/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/test/labels"
]

# Allowed classes (keep only stains (0) and holes (3))
allowed_classes = {'0', '3'}

# Function to filter labels
def filter_labels(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)

            # Read the file and filter lines
            with open(file_path, "r") as f:
                lines = f.readlines()

            # Keep only lines with class 0 or 3
            filtered_lines = [line for line in lines if line.split()[0] in allowed_classes]

            if filtered_lines:
                # Overwrite the file with only valid labels
                with open(file_path, "w") as f:
                    f.writelines(filtered_lines)
            else:
                # If no valid labels remain, delete the file
                os.remove(file_path)

# Apply the function to all label folders
for dir_path in label_dirs:
    filter_labels(dir_path)

print("Filtering completed. Only labels 0 (stains) and 3 (holes) are retained.")
```

```python
import os

# List of label directories
label_dirs = [
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/train/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/valid/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/test/labels"
]

# Function to update labels
def update_labels(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)

            # Read file content
            with open(file_path, "r") as f:
                lines = f.readlines()

            # Modify labels (convert 3 → 1)
            updated_lines = [line.replace("3 ", "1 ", 1) if line.startswith("3 ") else line for line in lines]

            # Write back updated labels
            with open(file_path, "w") as f:
                f.writelines(updated_lines)

# Apply the function to all label folders
for dir_path in label_dirs:
    update_labels(dir_path)

print("Label conversion completed: 3 → 1, 0 remains unchanged.")
```

```python
import os

# List of label directories
label_dirs = [
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/train/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/valid/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/test/labels"
]

# Function to update labels: convert 3 → 1, keep 0 and 2 unchanged
def update_labels(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)

            with open(file_path, "r") as f:
                lines = f.readlines()

            updated_lines = []
            for line in lines:
                parts = line.strip().split()
                if parts:
                    class_id = parts[0]
                    if class_id == '3':
                        parts[0] = '1'
                    updated_line = ' '.join(parts) + '\n'
                    updated_lines.append(updated_line)

            with open(file_path, "w") as f:
                f.writelines(updated_lines)

# Apply to all label folders
for dir_path in label_dirs:
    update_labels(dir_path)

print("✅ Label update completed: 3 → 1, 0 and 2 unchanged.")
```

```python
import os

# List of label directories
label_dirs = [
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/train/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/valid/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/test/labels"
]

# Allowed classes: 0 = stains, 3 = holes, 1 = tear cut
allowed_classes = {'0','1','3'}

# Function to filter labels
def filter_labels(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)

            # Read and filter lines
            with open(file_path, "r") as f:
                lines = f.readlines()

            # Keep only lines with allowed classes
            filtered_lines = [line for line in lines if line.split()[0] in allowed_classes]

            if filtered_lines:
                # Overwrite the file with the valid labels
                with open(file_path, "w") as f:
                    f.writelines(filtered_lines)
            else:
                # Delete the file if no valid labels remain
                os.remove(file_path)

# Apply the filtering function
for dir_path in label_dirs:
    filter_labels(dir_path)

print("✅ Filtering completed. Only classes 0 (stains), 1 (tear cut), and 3 (holes) are retained.")
```

#  Label update completed: 3 → 2, 0 and 1 unchanged.

```python
import os

# List of label directories
label_dirs = [
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/train/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/valid/labels",
    "/content/drive/MyDrive/mini project YOLO based/FD_Dataset/FD_Dataset/YOLO/test/labels"
]

# Function to update labels: convert 3 → 2, keep 0 and 1 unchanged
def update_labels(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)

            with open(file_path, "r") as f:
                lines = f.readlines()

            updated_lines = []
            for line in lines:
                parts = line.strip().split()
                if parts:
                    class_id = parts[0]
                    if class_id == '3':
                        parts[0] = '2'
                    updated_line = ' '.join(parts) + '\n'
                    updated_lines.append(updated_line)

            with open(file_path, "w") as f:
                f.writelines(updated_lines)

# Apply to all label folders
for dir_path in label_dirs:
    update_labels(dir_path)

print("✅ Label update completed: 3 → 2, 0 and 1 unchanged.")
```
